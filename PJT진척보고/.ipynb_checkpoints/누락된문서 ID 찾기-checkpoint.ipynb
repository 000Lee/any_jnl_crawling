```python
# ì—…ë¬´ê´€ë¦¬ ëˆ„ë½ëœ ë¬¸ì„œID ì°¾ê¸° + ìƒì„¸ ì •ë³´ í¬ë¡¤ë§
# txt íŒŒì¼ì˜ ë¬¸ì„œID ì¤‘ DBì— ì—†ëŠ” ê²ƒì„ ì°¾ê³ , ì›¹ì—ì„œ ìƒì„¸ ì •ë³´ ìˆ˜ì§‘

"""
 [ì „ì²´ íë¦„]
 task_ids.txtì—ì„œ ë¬¸ì„œID ëª©ë¡ ë¡œë“œ â†’ DB ë¹„êµ â†’ ëˆ„ë½ ID í™•ì¸ â†’ ì›¹ í¬ë¡¤ë§ â†’ ìƒì„¸ ì •ë³´ ìˆ˜ì§‘

 [1ë‹¨ê³„: ëˆ„ë½ëœ ë¬¸ì„œ ì°¾ê¸°]
 - txt íŒŒì¼ ì½ê¸°: task_ids.txtì—ì„œ ë¬¸ì„œID ëª©ë¡ ë¡œë“œ
 - DB ì¡°íšŒ: jnl_documents í…Œì´ë¸”ì˜ source_id ì¡°íšŒ
 - ì°¨ì§‘í•© ê³„ì‚°: txtì—ëŠ” ìˆëŠ”ë° DBì—ëŠ” ì—†ëŠ” ë¬¸ì„œID ì¶”ì¶œ
 - txt ì €ì¥: ëˆ„ë½ëœ IDë§Œ missing_task_ids.txtë¡œ ì €ì¥

 [2ë‹¨ê³„: ì›¹ì—ì„œ ìƒì„¸ ì •ë³´ í¬ë¡¤ë§]
 - ë¸Œë¼ìš°ì € ì‹¤í–‰ â†’ ìˆ˜ë™ ë¡œê·¸ì¸/ë©”ë‰´ ì§„ì…/ì¡°íšŒ
 - ê·¸ë¦¬ë“œ ìˆœíšŒí•˜ë©° ëˆ„ë½ ë¬¸ì„œë§Œ ìˆ˜ì§‘ (ì œëª©, ì‹œì‘ì¼, ì‘ì„±ì)
 - ëª¨ë“  ëˆ„ë½ ë¬¸ì„œë¥¼ ì°¾ìœ¼ë©´ ì¡°ê¸° ì¢…ë£Œ

 [3ë‹¨ê³„: ê²°ê³¼ ì €ì¥]
 - missing_task_ids.txt: ëˆ„ë½ëœ ë¬¸ì„œIDë§Œ
 - missing_tasks_detail.csv: ë¬¸ì„œID, ì œëª©, ì‹œì‘ì¼, ì‘ì„±ì
"""

import pymysql
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import csv
import re

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ğŸ”§ ì„¤ì • (ì—¬ê¸°ë§Œ ìˆ˜ì •í•˜ì„¸ìš”!)
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

BASE_URL = "http://office.anyfive.com/home/"

# DB ì„¤ì •
DB_CONFIG = {
    'host': 'localhost',
    'user': 'root',
    'password': '1234',
    'database': 'any_jnl'  # DBëª… í™•ì¸ í•„ìš”
}

# DB í…Œì´ë¸”ëª…
DB_TABLE = "jnl_documents"

# ê¸°ê°„ ì„¤ì • (í¬ë¡¤ë§ ì‹œ í•„í„°ìš©)
START_DATE = "2025-01-01"
END_DATE = "2025-12-31"

# ì…ë ¥ íŒŒì¼ (í¬ë¡¤ë§í•œ ë¬¸ì„œID ëª©ë¡)
INPUT_FILE = "task_ids_2025.txt"

# ì¶œë ¥ íŒŒì¼
OUTPUT_TXT = "missing_task_ids_2025.txt"        # ëˆ„ë½ëœ IDë§Œ
OUTPUT_CSV = "missing_tasks_detail_2025.csv"    # ìƒì„¸ ì •ë³´ í¬í•¨

# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
# ğŸ§ª í…ŒìŠ¤íŠ¸ ëª¨ë“œ (ì‹¤ì œ ìˆ˜ì§‘ ì‹œ ì•„ë˜ 2ì¤„ì„ ì£¼ì„ì²˜ë¦¬í•˜ì„¸ìš”)
# TEST_MODE = True
# TEST_EXCLUDE_IDS = ['3601017', '3408102', '3618779']  # DBì—ì„œ ì œì™¸í•  ID
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ğŸ“¦ ìœ í‹¸ë¦¬í‹° í•¨ìˆ˜
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

def norm_space(s):
    """ê³µë°± ì •ê·œí™”"""
    if not s:
        return ""
    s = s.replace("\xa0", " ")
    s = re.sub(r'[\t\r\n]+', ' ', s)
    s = re.sub(r'\s+', ' ', s)
    return s.strip()

def switch_into_frame_having(driver, selector_css, max_retries=3):
    """selectorê°€ ìˆëŠ” í”„ë ˆì„ìœ¼ë¡œ ì „í™˜"""
    for attempt in range(max_retries):
        driver.switch_to.default_content()
        
        try:
            WebDriverWait(driver, 5).until(
                EC.invisibility_of_element_located((By.CSS_SELECTOR, ".k-loading-mask"))
            )
        except:
            pass
        
        time.sleep(1)
        
        if driver.find_elements(By.CSS_SELECTOR, selector_css):
            print(f"  âœ“ ê·¸ë¦¬ë“œ ë°œê²¬ (ìµœìƒìœ„)")
            return True
        
        frames = driver.find_elements(By.CSS_SELECTOR, "iframe, frame")
        for idx, fr in enumerate(frames):
            try:
                driver.switch_to.default_content()
                driver.switch_to.frame(fr)
                if driver.find_elements(By.CSS_SELECTOR, selector_css):
                    print(f"  âœ“ ê·¸ë¦¬ë“œ ë°œê²¬ (í”„ë ˆì„ {idx})")
                    return True
            except:
                continue
        
        if attempt < max_retries - 1:
            print(f"  âš ï¸ ì¬ì‹œë„ {attempt + 1}/{max_retries}")
            time.sleep(2)
    
    return False

def wait_grid_stable(driver, timeout=10):
    """ê·¸ë¦¬ë“œ ì•ˆì •í™” ëŒ€ê¸°"""
    try:
        WebDriverWait(driver, timeout).until(
            EC.invisibility_of_element_located((By.CSS_SELECTOR, ".k-loading-mask"))
        )
        WebDriverWait(driver, timeout).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, "table[role='grid'] tbody tr"))
        )
        time.sleep(1)
        return True
    except:
        return False

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ğŸ•·ï¸ ê·¸ë¦¬ë“œ íŒŒì‹± (ëˆ„ë½ëœ ë¬¸ì„œë§Œ í•„í„°ë§)
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

def parse_grid_for_missing(driver, missing_ids):
    """í˜„ì¬ í˜ì´ì§€ì—ì„œ ëˆ„ë½ëœ ë¬¸ì„œë§Œ ì¶”ì¶œ
    
    ì¶”ì¶œ í•„ë“œ:
    - ë¬¸ì„œID: ì²´í¬ë°•ìŠ¤ value ë˜ëŠ” jnlInfoList.showGoì—ì„œ ì¶”ì¶œ
    - ì œëª©: <a class="grdTit ws_none"> íƒœê·¸ í…ìŠ¤íŠ¸
    - ì‹œì‘ì¼: data-field="work_strt_date" ì»¬ëŸ¼ (6ë²ˆì§¸ td)
    - ì‘ì„±ì: <span class="abcUsr"> íƒœê·¸ í…ìŠ¤íŠ¸
    """
    table = driver.find_element(By.CSS_SELECTOR, "table[role='grid']")
    rows = table.find_elements(By.CSS_SELECTOR, "tbody tr")
    
    results = []
    
    for tr in rows:
        tds = tr.find_elements(By.CSS_SELECTOR, "td")
        if not tds:
            continue
        
        # ë¬¸ì„œID ì¶”ì¶œ (ì²´í¬ë°•ìŠ¤ì—ì„œ)
        cb = tr.find_elements(By.CSS_SELECTOR, "input.jnlInfoListCheckBox")
        if cb:
            doc_id = cb[0].get_attribute("value").strip()
        else:
            # ëŒ€ì•ˆ: onclickì—ì„œ ì¶”ì¶œ
            link = tr.find_elements(By.CSS_SELECTOR, "a[onclick*='jnlInfoList.showGo']")
            if link:
                onclick = link[0].get_attribute("onclick") or ""
                match = re.search(r'jnlInfoList\.showGo\s*\(\s*this\s*,\s*(\d+)', onclick)
                doc_id = match.group(1) if match else ""
            else:
                continue
        
        if not doc_id:
            continue
        
        # ëˆ„ë½ëœ ë¬¸ì„œë§Œ ìˆ˜ì§‘
        if doc_id not in missing_ids:
            continue
        
        # ì œëª© ì¶”ì¶œ
        title_link = tr.find_elements(By.CSS_SELECTOR, "a.grdTit")
        title = norm_space(title_link[0].text) if title_link else ""
        
        # ì‹œì‘ì¼ ì¶”ì¶œ (6ë²ˆì§¸ td, index=5)
        start_date = ""
        if len(tds) > 5:
            start_date = norm_space(tds[5].text)
        
        # ì‘ì„±ì ì¶”ì¶œ
        writer_span = tr.find_elements(By.CSS_SELECTOR, "span.abcUsr")
        writer = norm_space(writer_span[0].text) if writer_span else ""
        
        item = {
            "ë¬¸ì„œID": doc_id,
            "ì œëª©": title,
            "ì‹œì‘ì¼": start_date,
            "ì‘ì„±ì": writer,
        }
        
        results.append(item)
    
    return results

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ğŸ”„ í˜ì´ì§•
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

def has_next_page(driver):
    """ë‹¤ìŒ í˜ì´ì§€ ì¡´ì¬ ì—¬ë¶€"""
    try:
        # ì˜ì–´ ë²„ì „
        next_btns = driver.find_elements(By.CSS_SELECTOR, 
            "a.k-pager-nav[aria-label='Go to the next page']:not(.k-state-disabled)")
        if next_btns:
            return True
        
        # í•œê¸€ ë²„ì „
        next_btns = driver.find_elements(By.CSS_SELECTOR, 
            "a.k-pager-nav[aria-label='ë‹¤ìŒ í˜ì´ì§€ë¡œ ì´ë™']:not(.k-state-disabled)")
        if next_btns:
            return True
        
        return False
    except:
        return False

def go_to_next_page(driver):
    """ë‹¤ìŒ í˜ì´ì§€ë¡œ ì´ë™"""
    try:
        # ì˜ì–´ ë²„ì „
        next_btn = driver.find_elements(By.CSS_SELECTOR,
            "a.k-pager-nav[aria-label='Go to the next page']:not(.k-state-disabled)")
        
        if not next_btn:
            # í•œê¸€ ë²„ì „
            next_btn = driver.find_elements(By.CSS_SELECTOR,
                "a.k-pager-nav[aria-label='ë‹¤ìŒ í˜ì´ì§€ë¡œ ì´ë™']:not(.k-state-disabled)")
        
        if next_btn:
            driver.execute_script("arguments[0].click();", next_btn[0])
            time.sleep(2)
            
            try:
                WebDriverWait(driver, 15).until(
                    EC.invisibility_of_element_located((By.CSS_SELECTOR, ".k-loading-mask"))
                )
            except:
                pass
            
            wait_grid_stable(driver)
            return True
        
        return False
    except Exception as e:
        print(f"  âŒ í˜ì´ì§€ ì´ë™ ì‹¤íŒ¨: {e}")
        return False

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ğŸš€ ë©”ì¸
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

def main():
    print("=" * 60)
    print("ğŸ” ì—…ë¬´ê´€ë¦¬ ëˆ„ë½ëœ ë¬¸ì„œID ì°¾ê¸° + ìƒì„¸ ì •ë³´ í¬ë¡¤ë§")
    print("=" * 60)
    
    # í…ŒìŠ¤íŠ¸ ëª¨ë“œ í‘œì‹œ
    if 'TEST_MODE' in globals() and TEST_MODE:
        print(f"âš ï¸  í…ŒìŠ¤íŠ¸ ëª¨ë“œ í™œì„±í™” (ì œì™¸ ID: {TEST_EXCLUDE_IDS})")
    
    # â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    # 1ë‹¨ê³„: txt vs DB ë¹„êµ
    # â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    print(f"\nğŸ“„ {INPUT_FILE} ì½ëŠ” ì¤‘...")
    with open(INPUT_FILE, "r", encoding="utf-8") as f:
        content = f.read().strip()
    
    txt_ids = set()
    for item in content.split(","):
        doc_id = item.strip().strip('"')
        if doc_id:
            txt_ids.add(doc_id)
    
    print(f"  âœ“ txt íŒŒì¼: {len(txt_ids)}ê±´")
    
    print(f"\nğŸ—„ï¸ DB ì¡°íšŒ ì¤‘... ({DB_TABLE})")
    conn = pymysql.connect(**DB_CONFIG)
    try:
        with conn.cursor() as cursor:
            cursor.execute(f"SELECT source_id FROM {DB_TABLE}")
            rows = cursor.fetchall()
            db_ids = set(str(row[0]) for row in rows if row[0])
    finally:
        conn.close()
    
    print(f"  âœ“ DB í…Œì´ë¸”: {len(db_ids)}ê±´")
    
    # â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    # ğŸ§ª í…ŒìŠ¤íŠ¸ ëª¨ë“œ: DBì—ì„œ íŠ¹ì • ID ì œì™¸ (ì‹¤ì œ ìˆ˜ì§‘ ì‹œ ì£¼ì„ì²˜ë¦¬)
    if 'TEST_MODE' in globals() and TEST_MODE:
        db_ids -= set(TEST_EXCLUDE_IDS)
        print(f"  âš ï¸ í…ŒìŠ¤íŠ¸: {len(TEST_EXCLUDE_IDS)}ê±´ ì œì™¸ë¨")
    # â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    
    # ì°¨ì§‘í•© ê³„ì‚°
    missing_ids = txt_ids - db_ids
    
    print(f"\n{'=' * 60}")
    print(f"ğŸ“Š ë¹„êµ ê²°ê³¼")
    print(f"{'=' * 60}")
    print(f"  â€¢ txt íŒŒì¼: {len(txt_ids)}ê±´")
    print(f"  â€¢ DB í…Œì´ë¸”: {len(db_ids)}ê±´")
    print(f"  â€¢ ëˆ„ë½ëœ ë¬¸ì„œ: {len(missing_ids)}ê±´")
    
    if not missing_ids:
        print(f"\nâœ… ëª¨ë“  ë¬¸ì„œê°€ DBì— ìˆìŠµë‹ˆë‹¤!")
        return
    
    # ëˆ„ë½ëœ IDë§Œ ì €ì¥ (ê¸°ì¡´ ê¸°ëŠ¥ ìœ ì§€)
    sorted_ids = sorted(missing_ids, key=lambda x: int(x) if x.isdigit() else 0, reverse=True)
    output_text = ",".join(f'"{doc_id}"' for doc_id in sorted_ids)
    with open(OUTPUT_TXT, "w", encoding="utf-8") as f:
        f.write(output_text)
    print(f"\nğŸ“„ ID ëª©ë¡ ì €ì¥: {OUTPUT_TXT}")
    
    # â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    # 2ë‹¨ê³„: ì›¹ì—ì„œ ìƒì„¸ ì •ë³´ í¬ë¡¤ë§
    # â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    print(f"\n{'=' * 60}")
    print(f"ğŸ•·ï¸ ëˆ„ë½ëœ ë¬¸ì„œ ìƒì„¸ ì •ë³´ í¬ë¡¤ë§")
    print(f"{'=' * 60}")
    
    # ë¸Œë¼ìš°ì € ì‹œì‘
    opts = webdriver.ChromeOptions()
    opts.add_argument("--start-maximized")
    opts.add_argument("--disable-popup-blocking")
    opts.add_argument("--disable-blink-features=AutomationControlled")
    opts.add_experimental_option("excludeSwitches", ["enable-automation"])
    
    driver = webdriver.Chrome(service=Service(), options=opts)
    
    try:
        driver.get(BASE_URL)
        time.sleep(1)
        
        print("\n" + "=" * 60)
        print("ğŸ“‹ ìˆ˜ë™ ì‘ì—…ì´ í•„ìš”í•©ë‹ˆë‹¤!")
        print("=" * 60)
        print("ë‹¤ìŒ ë‹¨ê³„ë¥¼ ì™„ë£Œí•œ í›„ Enterë¥¼ ëˆŒëŸ¬ì£¼ì„¸ìš”:\n")
        print("  1ï¸âƒ£  ë¡œê·¸ì¸")
        print("  2ï¸âƒ£  ì—…ë¬´ê´€ë¦¬ ë©”ë‰´ í´ë¦­")
        print("  3ï¸âƒ£  ì›í•˜ëŠ” í•˜ìœ„ ë©”ë‰´ ì„ íƒ (ì£¼ê°„ì—…ë¬´, ì—…ë¬´ì¼ì§€ ë“±)")
        print(f"  4ï¸âƒ£  í•„í„° ì„¤ì • (í•„ìš”ì‹œ)")
        print(f"  5ï¸âƒ£  ì‹œì‘ì¼ = {START_DATE}")
        print(f"  6ï¸âƒ£  ì¢…ë£Œì¼ = {END_DATE}")
        print("  7ï¸âƒ£  ì¡°íšŒ ë²„íŠ¼ í´ë¦­ â†’ ëª©ë¡ ë‚˜íƒ€ë‚¨ í™•ì¸")
        print("\nğŸ’¡ ëª©ë¡ì´ í™”ë©´ì— ë³´ì´ë©´ Enterë¥¼ ëˆ„ë¥´ì„¸ìš”!")
        print("=" * 60)
        input("\nâœ‹ ì¤€ë¹„ë˜ë©´ Enter... ")
        
        time.sleep(1)
        
        if not switch_into_frame_having(driver, "table[role='grid']"):
            print("âŒ ê·¸ë¦¬ë“œë¥¼ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤")
            # ëŒ€ì•ˆ ì…€ë ‰í„°ë¡œ ì¬ì‹œë„
            if not switch_into_frame_having(driver, "input.jnlInfoListCheckBox"):
                print("âŒ ì—…ë¬´ê´€ë¦¬ ëª©ë¡ì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤")
                return
        
        wait_grid_stable(driver)
        
        # í˜ì´ì§€ë³„ ìˆœíšŒí•˜ë©° ëˆ„ë½ëœ ë¬¸ì„œ ìˆ˜ì§‘
        found_docs = []
        found_ids = set()
        current_page = 1
        
        while True:
            print(f"\nğŸ“„ í˜ì´ì§€ {current_page} ìŠ¤ìº” ì¤‘...")
            
            if not switch_into_frame_having(driver, "table[role='grid']"):
                if not switch_into_frame_having(driver, "input.jnlInfoListCheckBox"):
                    print("âš ï¸ ê·¸ë¦¬ë“œ ì¬íƒìƒ‰ ì‹¤íŒ¨")
                    break
            
            wait_grid_stable(driver)
            
            # ëˆ„ë½ëœ ë¬¸ì„œë§Œ ì¶”ì¶œ
            page_docs = parse_grid_for_missing(driver, missing_ids)
            
            if page_docs:
                found_docs.extend(page_docs)
                for doc in page_docs:
                    found_ids.add(doc["ë¬¸ì„œID"])
                print(f"  âœ“ ëˆ„ë½ ë¬¸ì„œ {len(page_docs)}ê±´ ë°œê²¬ (ëˆ„ì : {len(found_docs)}ê±´)")
            else:
                print(f"  - ì´ í˜ì´ì§€ì— ëˆ„ë½ ë¬¸ì„œ ì—†ìŒ")
            
            # ëª¨ë‘ ì°¾ì•˜ìœ¼ë©´ ì¡°ê¸° ì¢…ë£Œ
            if found_ids == missing_ids:
                print(f"\nğŸ‰ ëª¨ë“  ëˆ„ë½ ë¬¸ì„œë¥¼ ì°¾ì•˜ìŠµë‹ˆë‹¤!")
                break
            
            # ë‹¤ìŒ í˜ì´ì§€
            if has_next_page(driver):
                if go_to_next_page(driver):
                    current_page += 1
                else:
                    break
            else:
                print(f"\nğŸ“„ ë§ˆì§€ë§‰ í˜ì´ì§€ ë„ë‹¬")
                break
        
        # â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
        # 3ë‹¨ê³„: ê²°ê³¼ ì¶œë ¥ ë° ì €ì¥
        # â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
        print(f"\n{'=' * 60}")
        print(f"ğŸ“Š í¬ë¡¤ë§ ê²°ê³¼")
        print(f"{'=' * 60}")
        print(f"  â€¢ ëˆ„ë½ëœ ë¬¸ì„œ: {len(missing_ids)}ê±´")
        print(f"  â€¢ ì°¾ì€ ë¬¸ì„œ: {len(found_docs)}ê±´")
        
        not_found = missing_ids - found_ids
        if not_found:
            print(f"  â€¢ ëª» ì°¾ì€ ë¬¸ì„œ: {len(not_found)}ê±´")
            print(f"    â†’ {list(not_found)[:10]}...")  # ì²˜ìŒ 10ê°œë§Œ í‘œì‹œ
        
        if found_docs:
            # CSV ì €ì¥
            with open(OUTPUT_CSV, "w", encoding="utf-8-sig", newline="") as f:
                writer = csv.DictWriter(f, fieldnames=["ë¬¸ì„œID", "ì œëª©", "ì‹œì‘ì¼", "ì‘ì„±ì"])
                writer.writeheader()
                writer.writerows(found_docs)
            
            print(f"\nâœ… ìƒì„¸ ì •ë³´ ì €ì¥: {OUTPUT_CSV}")
            
            # í™”ë©´ ì¶œë ¥
            print(f"\n{'â”€' * 100}")
            print(f"{'ë¬¸ì„œID':<12} {'ì œëª©':<50} {'ì‹œì‘ì¼':<12} {'ì‘ì„±ì':<10}")
            print(f"{'â”€' * 100}")
            for doc in found_docs[:20]:  # ì²˜ìŒ 20ê°œë§Œ í‘œì‹œ
                title = doc["ì œëª©"][:48] + ".." if len(doc["ì œëª©"]) > 50 else doc["ì œëª©"]
                print(f"{doc['ë¬¸ì„œID']:<12} {title:<50} {doc['ì‹œì‘ì¼']:<12} {doc['ì‘ì„±ì']:<10}")
            
            if len(found_docs) > 20:
                print(f"  ... ì™¸ {len(found_docs) - 20}ê±´")
            print(f"{'â”€' * 100}")
        
        print(f"\n{'=' * 60}")
        
    except Exception as e:
        print(f"\nâŒ ì˜¤ë¥˜ ë°œìƒ: {e}")
        import traceback
        traceback.print_exc()
    
    finally:
        input("\nâœ‹ ë¸Œë¼ìš°ì €ë¥¼ ë‹«ìœ¼ë ¤ë©´ Enterë¥¼ ëˆ„ë¥´ì„¸ìš”...")
        driver.quit()

if __name__ == "__main__":
    main()
```

```python
# ì—…ë¬´ê´€ë¦¬ ëˆ„ë½ëœ ë¬¸ì„œID ì°¾ê¸° + ìƒì„¸ ì •ë³´ í¬ë¡¤ë§
# txt íŒŒì¼ì˜ ë¬¸ì„œID ì¤‘ DBì— ì—†ëŠ” ê²ƒì„ ì°¾ê³ , ì›¹ì—ì„œ ìƒì„¸ ì •ë³´ ìˆ˜ì§‘

"""
 [ì „ì²´ íë¦„]
 task_ids.txtì—ì„œ ë¬¸ì„œID ëª©ë¡ ë¡œë“œ â†’ DB ë¹„êµ â†’ ëˆ„ë½ ID í™•ì¸ â†’ ì›¹ í¬ë¡¤ë§ â†’ ìƒì„¸ ì •ë³´ ìˆ˜ì§‘

 [1ë‹¨ê³„: ëˆ„ë½ëœ ë¬¸ì„œ ì°¾ê¸°]
 - txt íŒŒì¼ ì½ê¸°: task_ids.txtì—ì„œ ë¬¸ì„œID ëª©ë¡ ë¡œë“œ
 - DB ì¡°íšŒ: jnl_documents í…Œì´ë¸”ì˜ source_id ì¡°íšŒ
 - ì°¨ì§‘í•© ê³„ì‚°: txtì—ëŠ” ìˆëŠ”ë° DBì—ëŠ” ì—†ëŠ” ë¬¸ì„œID ì¶”ì¶œ
 - txt ì €ì¥: ëˆ„ë½ëœ IDë§Œ missing_task_ids.txtë¡œ ì €ì¥

 [2ë‹¨ê³„: ì›¹ì—ì„œ ìƒì„¸ ì •ë³´ í¬ë¡¤ë§]
 - ë¸Œë¼ìš°ì € ì‹¤í–‰ â†’ ìˆ˜ë™ ë¡œê·¸ì¸/ë©”ë‰´ ì§„ì…/ì¡°íšŒ
 - ê·¸ë¦¬ë“œ ìˆœíšŒí•˜ë©° ëˆ„ë½ ë¬¸ì„œë§Œ ìˆ˜ì§‘ (ì œëª©, ì‹œì‘ì¼, ì‘ì„±ì)
 - ëª¨ë“  ëˆ„ë½ ë¬¸ì„œë¥¼ ì°¾ìœ¼ë©´ ì¡°ê¸° ì¢…ë£Œ

 [3ë‹¨ê³„: ê²°ê³¼ ì €ì¥]
 - missing_task_ids.txt: ëˆ„ë½ëœ ë¬¸ì„œIDë§Œ
 - missing_tasks_detail.csv: ë¬¸ì„œID, ì œëª©, ì‹œì‘ì¼, ì‘ì„±ì
"""

import pymysql
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import csv
import re

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ğŸ”§ ì„¤ì • (ì—¬ê¸°ë§Œ ìˆ˜ì •í•˜ì„¸ìš”!)
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

BASE_URL = "http://office.anyfive.com/home/"

# DB ì„¤ì •
DB_CONFIG = {
    'host': 'localhost',
    'user': 'root',
    'password': '1234',
    'database': 'any_jnl'  # DBëª… í™•ì¸ í•„ìš”
}

# DB í…Œì´ë¸”ëª…
DB_TABLE = "jnl_documents"

# ê¸°ê°„ ì„¤ì • (í¬ë¡¤ë§ ì‹œ í•„í„°ìš©)
START_DATE = "2025-01-01"
END_DATE = "2025-12-31"

# ì…ë ¥ íŒŒì¼ (í¬ë¡¤ë§í•œ ë¬¸ì„œID ëª©ë¡)
INPUT_FILE = "task_ids_2025.txt"

# ì¶œë ¥ íŒŒì¼
OUTPUT_TXT = "missing_task_ids_2025.txt"        # ëˆ„ë½ëœ IDë§Œ
OUTPUT_CSV = "missing_tasks_detail_2025.csv"    # ìƒì„¸ ì •ë³´ í¬í•¨

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ğŸ“¦ ìœ í‹¸ë¦¬í‹° í•¨ìˆ˜
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

def norm_space(s):
    """ê³µë°± ì •ê·œí™”"""
    if not s:
        return ""
    s = s.replace("\xa0", " ")
    s = re.sub(r'[\t\r\n]+', ' ', s)
    s = re.sub(r'\s+', ' ', s)
    return s.strip()

def switch_into_frame_having(driver, selector_css, max_retries=3):
    """selectorê°€ ìˆëŠ” í”„ë ˆì„ìœ¼ë¡œ ì „í™˜"""
    for attempt in range(max_retries):
        driver.switch_to.default_content()
        
        try:
            WebDriverWait(driver, 5).until(
                EC.invisibility_of_element_located((By.CSS_SELECTOR, ".k-loading-mask"))
            )
        except:
            pass
        
        time.sleep(1)
        
        if driver.find_elements(By.CSS_SELECTOR, selector_css):
            print(f"  âœ“ ê·¸ë¦¬ë“œ ë°œê²¬ (ìµœìƒìœ„)")
            return True
        
        frames = driver.find_elements(By.CSS_SELECTOR, "iframe, frame")
        for idx, fr in enumerate(frames):
            try:
                driver.switch_to.default_content()
                driver.switch_to.frame(fr)
                if driver.find_elements(By.CSS_SELECTOR, selector_css):
                    print(f"  âœ“ ê·¸ë¦¬ë“œ ë°œê²¬ (í”„ë ˆì„ {idx})")
                    return True
            except:
                continue
        
        if attempt < max_retries - 1:
            print(f"  âš ï¸ ì¬ì‹œë„ {attempt + 1}/{max_retries}")
            time.sleep(2)
    
    return False

def wait_grid_stable(driver, timeout=10):
    """ê·¸ë¦¬ë“œ ì•ˆì •í™” ëŒ€ê¸°"""
    try:
        WebDriverWait(driver, timeout).until(
            EC.invisibility_of_element_located((By.CSS_SELECTOR, ".k-loading-mask"))
        )
        WebDriverWait(driver, timeout).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, "table[role='grid'] tbody tr"))
        )
        time.sleep(1)
        return True
    except:
        return False

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ğŸ•·ï¸ ê·¸ë¦¬ë“œ íŒŒì‹± (ëˆ„ë½ëœ ë¬¸ì„œë§Œ í•„í„°ë§)
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

def parse_grid_for_missing(driver, missing_ids):
    """í˜„ì¬ í˜ì´ì§€ì—ì„œ ëˆ„ë½ëœ ë¬¸ì„œë§Œ ì¶”ì¶œ
    
    ì¶”ì¶œ í•„ë“œ:
    - ë¬¸ì„œID: ì²´í¬ë°•ìŠ¤ value ë˜ëŠ” jnlInfoList.showGoì—ì„œ ì¶”ì¶œ
    - ì œëª©: <a class="grdTit ws_none"> íƒœê·¸ í…ìŠ¤íŠ¸
    - ì‹œì‘ì¼: data-field="work_strt_date" ì»¬ëŸ¼ (6ë²ˆì§¸ td)
    - ì‘ì„±ì: <span class="abcUsr"> íƒœê·¸ í…ìŠ¤íŠ¸
    """
    table = driver.find_element(By.CSS_SELECTOR, "table[role='grid']")
    rows = table.find_elements(By.CSS_SELECTOR, "tbody tr")
    
    results = []
    
    for tr in rows:
        tds = tr.find_elements(By.CSS_SELECTOR, "td")
        if not tds:
            continue
        
        # ë¬¸ì„œID ì¶”ì¶œ (ì²´í¬ë°•ìŠ¤ì—ì„œ)
        cb = tr.find_elements(By.CSS_SELECTOR, "input.jnlInfoListCheckBox")
        if cb:
            doc_id = cb[0].get_attribute("value").strip()
        else:
            # ëŒ€ì•ˆ: onclickì—ì„œ ì¶”ì¶œ
            link = tr.find_elements(By.CSS_SELECTOR, "a[onclick*='jnlInfoList.showGo']")
            if link:
                onclick = link[0].get_attribute("onclick") or ""
                match = re.search(r'jnlInfoList\.showGo\s*\(\s*this\s*,\s*(\d+)', onclick)
                doc_id = match.group(1) if match else ""
            else:
                continue
        
        if not doc_id:
            continue
        
        # ëˆ„ë½ëœ ë¬¸ì„œë§Œ ìˆ˜ì§‘
        if doc_id not in missing_ids:
            continue
        
        # ì œëª© ì¶”ì¶œ
        title_link = tr.find_elements(By.CSS_SELECTOR, "a.grdTit")
        title = norm_space(title_link[0].text) if title_link else ""
        
        # ì‹œì‘ì¼ ì¶”ì¶œ (6ë²ˆì§¸ td, index=5)
        start_date = ""
        if len(tds) > 5:
            start_date = norm_space(tds[5].text)
        
        # ì‘ì„±ì ì¶”ì¶œ
        writer_span = tr.find_elements(By.CSS_SELECTOR, "span.abcUsr")
        writer = norm_space(writer_span[0].text) if writer_span else ""
        
        item = {
            "ë¬¸ì„œID": doc_id,
            "ì œëª©": title,
            "ì‹œì‘ì¼": start_date,
            "ì‘ì„±ì": writer,
        }
        
        results.append(item)
    
    return results

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ğŸ”„ í˜ì´ì§•
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

def has_next_page(driver):
    """ë‹¤ìŒ í˜ì´ì§€ ì¡´ì¬ ì—¬ë¶€"""
    try:
        # ì˜ì–´ ë²„ì „
        next_btns = driver.find_elements(By.CSS_SELECTOR, 
            "a.k-pager-nav[aria-label='Go to the next page']:not(.k-state-disabled)")
        if next_btns:
            return True
        
        # í•œê¸€ ë²„ì „
        next_btns = driver.find_elements(By.CSS_SELECTOR, 
            "a.k-pager-nav[aria-label='ë‹¤ìŒ í˜ì´ì§€ë¡œ ì´ë™']:not(.k-state-disabled)")
        if next_btns:
            return True
        
        return False
    except:
        return False

def go_to_next_page(driver):
    """ë‹¤ìŒ í˜ì´ì§€ë¡œ ì´ë™"""
    try:
        # ì˜ì–´ ë²„ì „
        next_btn = driver.find_elements(By.CSS_SELECTOR,
            "a.k-pager-nav[aria-label='Go to the next page']:not(.k-state-disabled)")
        
        if not next_btn:
            # í•œê¸€ ë²„ì „
            next_btn = driver.find_elements(By.CSS_SELECTOR,
                "a.k-pager-nav[aria-label='ë‹¤ìŒ í˜ì´ì§€ë¡œ ì´ë™']:not(.k-state-disabled)")
        
        if next_btn:
            driver.execute_script("arguments[0].click();", next_btn[0])
            time.sleep(2)
            
            try:
                WebDriverWait(driver, 15).until(
                    EC.invisibility_of_element_located((By.CSS_SELECTOR, ".k-loading-mask"))
                )
            except:
                pass
            
            wait_grid_stable(driver)
            return True
        
        return False
    except Exception as e:
        print(f"  âŒ í˜ì´ì§€ ì´ë™ ì‹¤íŒ¨: {e}")
        return False

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ğŸš€ ë©”ì¸
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

def main():
    print("=" * 60)
    print("ğŸ” ì—…ë¬´ê´€ë¦¬ ëˆ„ë½ëœ ë¬¸ì„œID ì°¾ê¸° + ìƒì„¸ ì •ë³´ í¬ë¡¤ë§")
    print("=" * 60)
    
    # â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    # 1ë‹¨ê³„: txt vs DB ë¹„êµ
    # â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    print(f"\nğŸ“„ {INPUT_FILE} ì½ëŠ” ì¤‘...")
    with open(INPUT_FILE, "r", encoding="utf-8") as f:
        content = f.read().strip()
    
    txt_ids = set()
    for item in content.split(","):
        doc_id = item.strip().strip('"')
        if doc_id:
            txt_ids.add(doc_id)
    
    print(f"  âœ“ txt íŒŒì¼: {len(txt_ids)}ê±´")
    
    print(f"\nğŸ—„ï¸ DB ì¡°íšŒ ì¤‘... ({DB_TABLE})")
    conn = pymysql.connect(**DB_CONFIG)
    try:
        with conn.cursor() as cursor:
            cursor.execute(f"SELECT source_id FROM {DB_TABLE}")
            rows = cursor.fetchall()
            db_ids = set(str(row[0]) for row in rows if row[0])
    finally:
        conn.close()
    
    print(f"  âœ“ DB í…Œì´ë¸”: {len(db_ids)}ê±´")
    
    # ì°¨ì§‘í•© ê³„ì‚°
    missing_ids = txt_ids - db_ids
    
    print(f"\n{'=' * 60}")
    print(f"ğŸ“Š ë¹„êµ ê²°ê³¼")
    print(f"{'=' * 60}")
    print(f"  â€¢ txt íŒŒì¼: {len(txt_ids)}ê±´")
    print(f"  â€¢ DB í…Œì´ë¸”: {len(db_ids)}ê±´")
    print(f"  â€¢ ëˆ„ë½ëœ ë¬¸ì„œ: {len(missing_ids)}ê±´")
    
    if not missing_ids:
        print(f"\nâœ… ëª¨ë“  ë¬¸ì„œê°€ DBì— ìˆìŠµë‹ˆë‹¤!")
        return
    
    # ëˆ„ë½ëœ IDë§Œ ì €ì¥ (ê¸°ì¡´ ê¸°ëŠ¥ ìœ ì§€)
    sorted_ids = sorted(missing_ids, key=lambda x: int(x) if x.isdigit() else 0, reverse=True)
    output_text = ",".join(f'"{doc_id}"' for doc_id in sorted_ids)
    with open(OUTPUT_TXT, "w", encoding="utf-8") as f:
        f.write(output_text)
    print(f"\nğŸ“„ ID ëª©ë¡ ì €ì¥: {OUTPUT_TXT}")
    
    # â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    # 2ë‹¨ê³„: ì›¹ì—ì„œ ìƒì„¸ ì •ë³´ í¬ë¡¤ë§
    # â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    print(f"\n{'=' * 60}")
    print(f"ğŸ•·ï¸ ëˆ„ë½ëœ ë¬¸ì„œ ìƒì„¸ ì •ë³´ í¬ë¡¤ë§")
    print(f"{'=' * 60}")
    
    # ë¸Œë¼ìš°ì € ì‹œì‘
    opts = webdriver.ChromeOptions()
    opts.add_argument("--start-maximized")
    opts.add_argument("--disable-popup-blocking")
    opts.add_argument("--disable-blink-features=AutomationControlled")
    opts.add_experimental_option("excludeSwitches", ["enable-automation"])
    
    driver = webdriver.Chrome(service=Service(), options=opts)
    
    try:
        driver.get(BASE_URL)
        time.sleep(1)
        
        print("\n" + "=" * 60)
        print("ğŸ“‹ ìˆ˜ë™ ì‘ì—…ì´ í•„ìš”í•©ë‹ˆë‹¤!")
        print("=" * 60)
        print("ë‹¤ìŒ ë‹¨ê³„ë¥¼ ì™„ë£Œí•œ í›„ Enterë¥¼ ëˆŒëŸ¬ì£¼ì„¸ìš”:\n")
        print("  1ï¸âƒ£  ë¡œê·¸ì¸")
        print("  2ï¸âƒ£  ì—…ë¬´ê´€ë¦¬ ë©”ë‰´ í´ë¦­")
        print("  3ï¸âƒ£  ì›í•˜ëŠ” í•˜ìœ„ ë©”ë‰´ ì„ íƒ (ì£¼ê°„ì—…ë¬´, ì—…ë¬´ì¼ì§€ ë“±)")
        print(f"  4ï¸âƒ£  í•„í„° ì„¤ì • (í•„ìš”ì‹œ)")
        print(f"  5ï¸âƒ£  ì‹œì‘ì¼ = {START_DATE}")
        print(f"  6ï¸âƒ£  ì¢…ë£Œì¼ = {END_DATE}")
        print("  7ï¸âƒ£  ì¡°íšŒ ë²„íŠ¼ í´ë¦­ â†’ ëª©ë¡ ë‚˜íƒ€ë‚¨ í™•ì¸")
        print("\nğŸ’¡ ëª©ë¡ì´ í™”ë©´ì— ë³´ì´ë©´ Enterë¥¼ ëˆ„ë¥´ì„¸ìš”!")
        print("=" * 60)
        input("\nâœ‹ ì¤€ë¹„ë˜ë©´ Enter... ")
        
        time.sleep(1)
        
        if not switch_into_frame_having(driver, "table[role='grid']"):
            print("âŒ ê·¸ë¦¬ë“œë¥¼ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤")
            # ëŒ€ì•ˆ ì…€ë ‰í„°ë¡œ ì¬ì‹œë„
            if not switch_into_frame_having(driver, "input.jnlInfoListCheckBox"):
                print("âŒ ì—…ë¬´ê´€ë¦¬ ëª©ë¡ì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤")
                return
        
        wait_grid_stable(driver)
        
        # í˜ì´ì§€ë³„ ìˆœíšŒí•˜ë©° ëˆ„ë½ëœ ë¬¸ì„œ ìˆ˜ì§‘
        found_docs = []
        found_ids = set()
        current_page = 1
        
        while True:
            print(f"\nğŸ“„ í˜ì´ì§€ {current_page} ìŠ¤ìº” ì¤‘...")
            
            if not switch_into_frame_having(driver, "table[role='grid']"):
                if not switch_into_frame_having(driver, "input.jnlInfoListCheckBox"):
                    print("âš ï¸ ê·¸ë¦¬ë“œ ì¬íƒìƒ‰ ì‹¤íŒ¨")
                    break
            
            wait_grid_stable(driver)
            
            # ëˆ„ë½ëœ ë¬¸ì„œë§Œ ì¶”ì¶œ
            page_docs = parse_grid_for_missing(driver, missing_ids)
            
            if page_docs:
                found_docs.extend(page_docs)
                for doc in page_docs:
                    found_ids.add(doc["ë¬¸ì„œID"])
                print(f"  âœ“ ëˆ„ë½ ë¬¸ì„œ {len(page_docs)}ê±´ ë°œê²¬ (ëˆ„ì : {len(found_docs)}ê±´)")
            else:
                print(f"  - ì´ í˜ì´ì§€ì— ëˆ„ë½ ë¬¸ì„œ ì—†ìŒ")
            
            # ëª¨ë‘ ì°¾ì•˜ìœ¼ë©´ ì¡°ê¸° ì¢…ë£Œ
            if found_ids == missing_ids:
                print(f"\nğŸ‰ ëª¨ë“  ëˆ„ë½ ë¬¸ì„œë¥¼ ì°¾ì•˜ìŠµë‹ˆë‹¤!")
                break
            
            # ë‹¤ìŒ í˜ì´ì§€
            if has_next_page(driver):
                if go_to_next_page(driver):
                    current_page += 1
                else:
                    break
            else:
                print(f"\nğŸ“„ ë§ˆì§€ë§‰ í˜ì´ì§€ ë„ë‹¬")
                break
        
        # â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
        # 3ë‹¨ê³„: ê²°ê³¼ ì¶œë ¥ ë° ì €ì¥
        # â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
        print(f"\n{'=' * 60}")
        print(f"ğŸ“Š í¬ë¡¤ë§ ê²°ê³¼")
        print(f"{'=' * 60}")
        print(f"  â€¢ ëˆ„ë½ëœ ë¬¸ì„œ: {len(missing_ids)}ê±´")
        print(f"  â€¢ ì°¾ì€ ë¬¸ì„œ: {len(found_docs)}ê±´")
        
        not_found = missing_ids - found_ids
        if not_found:
            print(f"  â€¢ ëª» ì°¾ì€ ë¬¸ì„œ: {len(not_found)}ê±´")
            print(f"    â†’ {list(not_found)[:10]}...")  # ì²˜ìŒ 10ê°œë§Œ í‘œì‹œ
        
        if found_docs:
            # CSV ì €ì¥
            with open(OUTPUT_CSV, "w", encoding="utf-8-sig", newline="") as f:
                writer = csv.DictWriter(f, fieldnames=["ë¬¸ì„œID", "ì œëª©", "ì‹œì‘ì¼", "ì‘ì„±ì"])
                writer.writeheader()
                writer.writerows(found_docs)
            
            print(f"\nâœ… ìƒì„¸ ì •ë³´ ì €ì¥: {OUTPUT_CSV}")
            
            # í™”ë©´ ì¶œë ¥
            print(f"\n{'â”€' * 100}")
            print(f"{'ë¬¸ì„œID':<12} {'ì œëª©':<50} {'ì‹œì‘ì¼':<12} {'ì‘ì„±ì':<10}")
            print(f"{'â”€' * 100}")
            for doc in found_docs[:20]:  # ì²˜ìŒ 20ê°œë§Œ í‘œì‹œ
                title = doc["ì œëª©"][:48] + ".." if len(doc["ì œëª©"]) > 50 else doc["ì œëª©"]
                print(f"{doc['ë¬¸ì„œID']:<12} {title:<50} {doc['ì‹œì‘ì¼']:<12} {doc['ì‘ì„±ì']:<10}")
            
            if len(found_docs) > 20:
                print(f"  ... ì™¸ {len(found_docs) - 20}ê±´")
            print(f"{'â”€' * 100}")
        
        print(f"\n{'=' * 60}")
        
    except Exception as e:
        print(f"\nâŒ ì˜¤ë¥˜ ë°œìƒ: {e}")
        import traceback
        traceback.print_exc()
    
    finally:
        input("\nâœ‹ ë¸Œë¼ìš°ì €ë¥¼ ë‹«ìœ¼ë ¤ë©´ Enterë¥¼ ëˆ„ë¥´ì„¸ìš”...")
        driver.quit()

if __name__ == "__main__":
    main()
```
