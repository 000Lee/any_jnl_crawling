```python
# ì—…ë¬´ê´€ë¦¬ ë¬¸ì„œID ì¶”ì¶œê¸°
# íŠ¹ì • ê¸°ê°„ì˜ ë¬¸ì„œIDë§Œ í…ìŠ¤íŠ¸ íŒŒì¼ë¡œ ì¶”ì¶œ

"""
 [ê¸°ëŠ¥]
 ì—…ë¬´ê´€ë¦¬ ì‹œìŠ¤í…œì—ì„œ íŠ¹ì • ê¸°ê°„ì˜ ë¬¸ì„œIDë§Œ ì¶”ì¶œí•˜ì—¬ txt íŒŒì¼ë¡œ ì €ì¥

 [ì¶œë ¥ í˜•ì‹]
 "3630714","3630715","3630716" (í•œ ì¤„ì— ìŒë”°ì˜´í‘œ+ì‰¼í‘œ êµ¬ë¶„)

 [ë™ì‘ ìˆœì„œ]
 1. ë¸Œë¼ìš°ì € ì‹¤í–‰
 2. ìˆ˜ë™ ì‘ì—… ëŒ€ê¸° (ë¡œê·¸ì¸ â†’ ì—…ë¬´ê´€ë¦¬ ë©”ë‰´ â†’ í•„í„° ì„¤ì • â†’ ì¡°íšŒ)
 3. ê·¸ë¦¬ë“œì—ì„œ ë¬¸ì„œID ì¶”ì¶œ (onclickì˜ jnlInfoList.showGoì—ì„œ ì¶”ì¶œ)
 4. ë‹¤ìŒ í˜ì´ì§€ë¡œ ì´ë™í•˜ë©° ë°˜ë³µ
 5. txt íŒŒì¼ë¡œ ì €ì¥

 [ì„¤ì •]
 - START_DATE / END_DATE: ì¡°íšŒ ê¸°ê°„
 - OUTPUT_FILE: ì¶œë ¥ íŒŒì¼ëª…
"""

from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import re

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ğŸ”§ ì„¤ì • (ì—¬ê¸°ë§Œ ìˆ˜ì •í•˜ì„¸ìš”!)
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

BASE_URL = "http://office.anyfive.com/home/"

# ê¸°ê°„ ì„¤ì •
START_DATE = "2016-01-01"  # ì—¬ê¸°ë¥¼ ìˆ˜ì •í•˜ì„¸ìš” - ì‹œì‘ì¼
END_DATE = "2025-12-31"    # ì—¬ê¸°ë¥¼ ìˆ˜ì •í•˜ì„¸ìš” - ì¢…ë£Œì¼

# ì¶œë ¥ íŒŒì¼ëª…
OUTPUT_FILE = "task_ids_2025.txt"  # ì—¬ê¸°ë¥¼ ìˆ˜ì •í•˜ì„¸ìš”

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ğŸ“¦ ìœ í‹¸ë¦¬í‹° í•¨ìˆ˜
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

def switch_into_frame_having(driver, selector_css, max_retries=3):
    """selectorê°€ ìˆëŠ” í”„ë ˆì„ìœ¼ë¡œ ì „í™˜"""
    for attempt in range(max_retries):
        driver.switch_to.default_content()
        
        # ë¡œë”© ë§ˆìŠ¤í¬ ëŒ€ê¸°
        try:
            WebDriverWait(driver, 5).until(
                EC.invisibility_of_element_located((By.CSS_SELECTOR, ".k-loading-mask"))
            )
        except:
            pass
        
        time.sleep(1)
        
        # ìµœìƒìœ„ í™•ì¸
        if driver.find_elements(By.CSS_SELECTOR, selector_css):
            print(f"  âœ“ ê·¸ë¦¬ë“œ ë°œê²¬ (ìµœìƒìœ„)")
            return True
        
        # í”„ë ˆì„ ë‚´ë¶€ íƒìƒ‰
        frames = driver.find_elements(By.CSS_SELECTOR, "iframe, frame")
        for idx, fr in enumerate(frames):
            try:
                driver.switch_to.default_content()
                driver.switch_to.frame(fr)
                if driver.find_elements(By.CSS_SELECTOR, selector_css):
                    print(f"  âœ“ ê·¸ë¦¬ë“œ ë°œê²¬ (í”„ë ˆì„ {idx})")
                    return True
            except:
                continue
        
        if attempt < max_retries - 1:
            print(f"  âš ï¸ ì¬ì‹œë„ {attempt + 1}/{max_retries}")
            time.sleep(2)
    
    return False

def wait_grid_stable(driver, timeout=10):
    """ê·¸ë¦¬ë“œ ì•ˆì •í™” ëŒ€ê¸°"""
    try:
        WebDriverWait(driver, timeout).until(
            EC.invisibility_of_element_located((By.CSS_SELECTOR, ".k-loading-mask"))
        )
        # ì—…ë¬´ê´€ë¦¬ í…Œì´ë¸” ëŒ€ê¸°
        WebDriverWait(driver, timeout).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, "table tbody tr"))
        )
        time.sleep(1)
        return True
    except:
        return False

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ğŸ•·ï¸ ë¬¸ì„œID ì¶”ì¶œ
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

def extract_task_ids(driver):
    """í˜„ì¬ í˜ì´ì§€ì˜ ì—…ë¬´ê´€ë¦¬ ë¬¸ì„œID ì¶”ì¶œ
    
    onclick="jnlInfoList.showGo(this,3630714,1,true,'null')" ì—ì„œ ID ì¶”ì¶œ
    """
    doc_ids = []
    
    # jnlInfoList.showGoê°€ ìˆëŠ” ë§í¬ ì°¾ê¸°
    links = driver.find_elements(By.CSS_SELECTOR, "a[onclick*='jnlInfoList.showGo']")
    
    for link in links:
        try:
            onclick = link.get_attribute("onclick")
            if onclick:
                # jnlInfoList.showGo(this,3630714,... ì—ì„œ ìˆ«ì ì¶”ì¶œ
                match = re.search(r'jnlInfoList\.showGo\s*\(\s*this\s*,\s*(\d+)', onclick)
                if match:
                    doc_id = match.group(1)
                    if doc_id and doc_id not in doc_ids:
                        doc_ids.append(doc_id)
        except Exception as e:
            continue
    
    return doc_ids

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ğŸ”„ í˜ì´ì§•
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

def has_next_page(driver):
    """ë‹¤ìŒ í˜ì´ì§€ ì¡´ì¬ ì—¬ë¶€"""
    try:
        # Kendo í˜ì´ì € ìŠ¤íƒ€ì¼ (ì˜ì–´)
        next_btns = driver.find_elements(By.CSS_SELECTOR, 
            "a.k-pager-nav[aria-label='Go to the next page']:not(.k-state-disabled)")
        if next_btns:
            return True
        
        # Kendo í˜ì´ì € ìŠ¤íƒ€ì¼ (í•œê¸€)
        next_btns = driver.find_elements(By.CSS_SELECTOR, 
            "a.k-pager-nav[aria-label='ë‹¤ìŒ í˜ì´ì§€ë¡œ ì´ë™']:not(.k-state-disabled)")
        if next_btns:
            return True
            
        return False
    except:
        return False

def go_to_next_page(driver):
    """ë‹¤ìŒ í˜ì´ì§€ë¡œ ì´ë™"""
    try:
        # Kendo ìŠ¤íƒ€ì¼ (ì˜ì–´)
        next_btn = driver.find_elements(By.CSS_SELECTOR,
            "a.k-pager-nav[aria-label='Go to the next page']:not(.k-state-disabled)")
        
        if not next_btn:
            # Kendo ìŠ¤íƒ€ì¼ (í•œê¸€)
            next_btn = driver.find_elements(By.CSS_SELECTOR,
                "a.k-pager-nav[aria-label='ë‹¤ìŒ í˜ì´ì§€ë¡œ ì´ë™']:not(.k-state-disabled)")
        
        if next_btn:
            driver.execute_script("arguments[0].click();", next_btn[0])
            time.sleep(2)
            
            # ë¡œë”© ëŒ€ê¸°
            try:
                WebDriverWait(driver, 15).until(
                    EC.invisibility_of_element_located((By.CSS_SELECTOR, ".k-loading-mask"))
                )
            except:
                pass
            
            wait_grid_stable(driver)
            return True
        
        return False
    except Exception as e:
        print(f"  âŒ í˜ì´ì§€ ì´ë™ ì‹¤íŒ¨: {e}")
        return False

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ğŸš€ ë©”ì¸
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

def main():
    print("=" * 60)
    print("ğŸ” ì—…ë¬´ê´€ë¦¬ ë¬¸ì„œID ì¶”ì¶œê¸°")
    print(f"ğŸ“… ê¸°ê°„: {START_DATE} ~ {END_DATE}")
    print(f"ğŸ“„ ì¶œë ¥: {OUTPUT_FILE}")
    print("=" * 60)
    
    # ë¸Œë¼ìš°ì € ì‹œì‘
    opts = webdriver.ChromeOptions()
    opts.add_argument("--start-maximized")
    opts.add_argument("--disable-popup-blocking")
    opts.add_argument("--disable-blink-features=AutomationControlled")
    opts.add_experimental_option("excludeSwitches", ["enable-automation"])
    
    driver = webdriver.Chrome(service=Service(), options=opts)
    
    try:
        driver.get(BASE_URL)
        time.sleep(1)
        
        print("\n" + "=" * 60)
        print("ğŸ“‹ ìˆ˜ë™ ì‘ì—…ì´ í•„ìš”í•©ë‹ˆë‹¤!")
        print("=" * 60)
        print("ë‹¤ìŒ ë‹¨ê³„ë¥¼ ì™„ë£Œí•œ í›„ Enterë¥¼ ëˆŒëŸ¬ì£¼ì„¸ìš”:\n")
        print("  1ï¸âƒ£  ë¡œê·¸ì¸")
        print("  2ï¸âƒ£  ì—…ë¬´ê´€ë¦¬ ë©”ë‰´ í´ë¦­")
        print("  3ï¸âƒ£  ì›í•˜ëŠ” í•˜ìœ„ ë©”ë‰´ ì„ íƒ (ì£¼ê°„ì—…ë¬´, ì—…ë¬´ì¼ì§€ ë“±)")
        print(f"  4ï¸âƒ£  í•„í„° ì„¤ì • (í•„ìš”ì‹œ)")
        print(f"  5ï¸âƒ£  ì‹œì‘ì¼ = {START_DATE}")
        print(f"  6ï¸âƒ£  ì¢…ë£Œì¼ = {END_DATE}")
        print("  7ï¸âƒ£  ì¡°íšŒ ë²„íŠ¼ í´ë¦­ â†’ ëª©ë¡ ë‚˜íƒ€ë‚¨ í™•ì¸")
        print("\nğŸ’¡ ëª©ë¡ì´ í™”ë©´ì— ë³´ì´ë©´ Enterë¥¼ ëˆ„ë¥´ì„¸ìš”!")
        print("=" * 60)
        input("\nâœ‹ ì¤€ë¹„ë˜ë©´ Enter... ")
        
        time.sleep(1)
        
        # ê·¸ë¦¬ë“œë¡œ ì „í™˜ (jnlInfoList.showGo ë§í¬ê°€ ìˆëŠ” ê³³)
        if not switch_into_frame_having(driver, "a[onclick*='jnlInfoList.showGo']"):
            print("âŒ ì—…ë¬´ê´€ë¦¬ ëª©ë¡ì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤")
            print("ğŸ’¡ ë‹¤ë¥¸ ì…€ë ‰í„°ë¡œ ì¬ì‹œë„...")
            # ì¼ë°˜ í…Œì´ë¸”ë¡œ ì¬ì‹œë„
            if not switch_into_frame_having(driver, "table tbody tr"):
                print("âŒ í…Œì´ë¸”ì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤")
                return
        
        wait_grid_stable(driver)
        
        # ì „ì²´ ë¬¸ì„œID ìˆ˜ì§‘
        all_doc_ids = []
        current_page = 1
        
        while True:
            print(f"\nğŸ“„ í˜ì´ì§€ {current_page} ì²˜ë¦¬ ì¤‘...")
            
            # ê·¸ë¦¬ë“œ ì¬í™•ì¸
            switch_into_frame_having(driver, "a[onclick*='jnlInfoList.showGo']")
            wait_grid_stable(driver)
            
            # ë¬¸ì„œID ì¶”ì¶œ
            page_ids = extract_task_ids(driver)
            
            if not page_ids:
                print("âš ï¸ ë°ì´í„°ê°€ ì—†ìŠµë‹ˆë‹¤")
                break
            
            all_doc_ids.extend(page_ids)
            print(f"  âœ“ {len(page_ids)}ê±´ ì¶”ì¶œ (ëˆ„ì : {len(all_doc_ids)}ê±´)")
            
            # ë‹¤ìŒ í˜ì´ì§€
            if has_next_page(driver):
                if go_to_next_page(driver):
                    current_page += 1
                else:
                    break
            else:
                print(f"\nğŸ‰ ë§ˆì§€ë§‰ í˜ì´ì§€ ë„ë‹¬!")
                break
        
        # txt íŒŒì¼ ì €ì¥
        if all_doc_ids:
            output_text = ",".join(f'"{doc_id}"' for doc_id in all_doc_ids)
            with open(OUTPUT_FILE, "w", encoding="utf-8") as f:
                f.write(output_text)
            
            print(f"\n{'=' * 60}")
            print(f"âœ… ì™„ë£Œ!")
            print(f"{'=' * 60}")
            print(f"ğŸ“Š ì´ {len(all_doc_ids)}ê±´ ì¶”ì¶œ")
            print(f"ğŸ“„ ì €ì¥: {OUTPUT_FILE}")
            print(f"{'=' * 60}")
        else:
            print("\nâš ï¸ ì¶”ì¶œëœ ë¬¸ì„œIDê°€ ì—†ìŠµë‹ˆë‹¤")
        
    except Exception as e:
        print(f"\nâŒ ì˜¤ë¥˜ ë°œìƒ: {e}")
    
    finally:
        input("\nâœ‹ ë¸Œë¼ìš°ì €ë¥¼ ë‹«ìœ¼ë ¤ë©´ Enterë¥¼ ëˆ„ë¥´ì„¸ìš”...")
        driver.quit()

if __name__ == "__main__":
    main()
```
